```python
import numpy as np

def clean_date(item):
    print('Begins with \n', item)
    
    dop = str(item)
    if dop == "nan" or dop == 'NULL':
        return np.nan
    if dop[0]=='[':
        return np.nan
    for i in ('-', ',', '['):
        if i in dop:
            loc = dop.find(i)
            dop = dop[:loc]
    print('End')
    return dop
```

```python
df['Date of Publication'] = df['Date of Publication'].apply(clean_date)
```
